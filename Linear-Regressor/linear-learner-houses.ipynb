# ConfiguraÃ§Ãµes SageMaker

```python
import boto3
import sagemaker
from sagemaker import Session
```

```python
# iniciando a sessao e selecionando o bucket ja existente
sessao_boto = boto3.Session(region_name='sa-east-1')
session = sagemaker.Session(sessao_boto)
bucket = 'jescursoawssagemaker'
subpasta_modelo = 'modelos/house-prices/linear-learner'
subpasta_dataset = 'datasets/house-prices'
```

```python
role = sagemaker.get_execution_role()
role
```

```python
# converter nossa base de dados atual (numpy) para bytes pq o aws pede nesse formato
import io
import sagemaker.amazon.common as smac 

buffer = io.BytesIO()
smac.write_numpy_to_dense_tensor(buffer,X_train,y_train)
buffer.seek(0) # coloca a base de dados na posiÃ§Ã£o inicial
```

```python
# escrevendo nossa base de treino (buffet) no bucket com subpastas 
import os 
key = 'houses-train-data' # nome que vamos dar ao conjunto de treino
boto3.resource('s3').Bucket(bucket).Object(os.path.join(subpasta_dataset,'train',key)).upload_fileobj(buffer)
s3_train_data = 's3://{}/{}/train/{}'.format(bucket,subpasta_dataset,key)
print('LocalizaÃ§Ã£o da base de treinamento:{}'.format(s3_train_data))
```

```python
# instanciando onde sera salvo o modelo
output_location = 's3://{}/{}/output'.format(bucket,subpasta_modelo)
print('Modelo final serÃ¡ salvo em {}'.format(output_location))
```

# CriaÃ§Ã£o e treinamento do Linear Learner

```python
# documentacao util
# https://docs.aws.amazon.com/sagemaker/latest/dg/ecr-us-east-1.html
# https://docs.aws.amazon.com/pt_br/AWSEC2/latest/UserGuide/instance-types.html
```

```python
# instanciando qual modelo vamos utilizar em um container
# sobre o algoritmo: https://docs.aws.amazon.com/sagemaker/latest/dg/linear-learner.html
container = sagemaker.image_uris.retrieve(framework='linear-learner', region=boto3.Session().region_name)
```

```python
boto3.session.Session().region_name
```

```python
linear = sagemaker.estimator.Estimator(image_uri = container,
                                      role = role, 
                                      instance_count = 1, # vamos usar sÃ³ 1 servidor
                                      instance_type = 'ml.m4.xlarge', # instancia do EC2
                                      output_path = output_location, # local onde modelo sera salvo
                                      sagemaker_session = session # sessao que ja criamos para usar o SM 
                                      )

# documentacao: https://sagemaker.readthedocs.io/en/stable/api/training/estimators.html

# # para reduzir o custo pago na amazon, acrescentamos um parametro chamado use_stop_instance=True, pois ele aloca mÃ¡quinas virtuais que estÃ£o "paradas"
# linear = sagemaker.estimator.Estimator(image_uri = container,
#                                        role = role,
#                                        instance_count = 1,
#                                        instance_type = 'ml.m4.xlarge',
#                                        output_path = output_location,
#                                        sagemaker_session = session,
#                                        use_stop_instances = True)
```

```python
# criando os parametros do modelo
linear.set_hyperparameters(feature_dim = 16, # numero de features do vetor X
                          predictor_type = 'regressor',
                          num_models = 16, # validacao cruzada com 16 modelos
                          )

# documentacao : https://docs.aws.amazon.com/sagemaker/latest/dg/ll_hyperparameters.html
```

```python
linear.fit({'train': s3_train_data})
```

# Deploy, previsÃµes e avaliaÃ§Ã£o

```python
# https://aws.amazon.com/pt/ec2/instance-types/inf1/
linear_regressor = linear.deploy(initial_instance_count = 1, instance_type = 'ml.m4.xlarge')
```

```python
from sagemaker.serializers import CSVSerializer
from sagemaker.deserializers import JSONDeserializer

linear_regressor.serializer = CSVSerializer()
linear_regressor.deserializer = JSONDeserializer()
```

```python
y_pred = linear_regressor.predict(X_test)
```

```python
type(y_pred)
```

```python
previsoes = np.array([r['score'] for r in y_pred['predictions']])
```

```python
type(previsoes)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
R2  = r2_score(y_test,previsoes).round(4)
EQM = mean_squared_error(y_test,previsoes).round(4) #mais proximo de zero, melhor
REQM  = np.sqrt(EQM).round(4) #calcula a distancia de um ponto em relaÃ§Ã£o a linha, logo verifica a concentraÃ§Ã£o dos dados proximo a linha ajustada
```

```python
print('O valor de RÂ² para esse modelo RF Ã© {}'.format(R2.round(4)))
print('O valor de EQM para esse modelo RF Ã© {}'.format(EQM))
print('O valor de REQM para esse modelo RF Ã© {}'.format(REQM))
```
